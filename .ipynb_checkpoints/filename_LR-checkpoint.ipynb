```python
import os
#코드 순서 
#RL부여 한번 실행 (이미지파일 및 json파일명에 R,L이 붙게 됨)
#이름이 바뀐 이미지가 담긴 train 폴더를 train_LR로 폴더명을 수동으로 바꿔주기
#base_origin_LR 두번 실행 (left로한번, right로 한번 (중간에 dataset부분 코드 변경해서 left한번,모델 저장 이름 바꿔서 right한번씩 총 두번돌려야함))
#base_origin_LR을 통해 생긴 모델 2개 경로를 inferece_LR에 넣고 실행 


# 작업 디렉토리 설정
base_path = '/opt/ml/input/data/train/DCM'
jsonbase_path='/opt/ml/input/data/train/outputs_json'
# 524개 폴더에 대해 작업 수행
for i in range(2, 549): #ID001은 왜인지 모르겠는데 되어있음 
    folder_name = f'ID{i:03d}'  # 폴더 이름 형식 지정
    folder_path = os.path.join(base_path, folder_name)
    jsonfolder_path=os.path.join(jsonbase_path, folder_name)
    # 폴더가 존재하는지 확인
    if os.path.exists(folder_path):
        # 폴더 내 파일 목록 확인
        file_list = os.listdir(folder_path)

        # 이미지 파일 목록 필터링
        image_files = [file for file in file_list if file.startswith('image')]

        # 이미지 파일이 2개 이상인 경우에만 작업 수행
        if len(image_files) >= 2:
            # 이미지 파일 이름 정렬하여 순서대로 처리
            sorted_images = sorted(image_files, key=lambda x: int(x[5:-4]))

            # 첫 번째 이미지 파일 이름에 '_R' 추가
            first_image = sorted_images[0]
            new_name_r = first_image[:-4] + '_R' + first_image[-4:]
            os.rename(os.path.join(folder_path, first_image), os.path.join(folder_path, new_name_r))
            os.rename(os.path.join(jsonfolder_path, first_image[:-4]+'.json'), os.path.join(jsonfolder_path, new_name_r[:-4]+'.json'))

            # 두 번째 이미지 파일 이름에 '_L' 추가
            second_image = sorted_images[1]
            new_name_l = second_image[:-4] + '_L' + second_image[-4:]
            os.rename(os.path.join(folder_path, second_image), os.path.join(folder_path, new_name_l))
            os.rename(os.path.join(jsonfolder_path, second_image[:-4]+'.json'), os.path.join(jsonfolder_path, new_name_l[:-4]+'.json'))
    else:
        #print(f'폴더 {folder_name}이(가) 존재하지 않습니다.')
        continue
```
